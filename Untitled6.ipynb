```python
# -*- coding: utf-8 -*-
"""faktorial_rekursif.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/13l7BNax1z2J5x9OqWGEelSWYJc9l47-E
"""

def faktorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * faktorial(n - 1)

# Input dari pengguna
try:
    angka = int(input("Masukkan bilangan bulat positif: "))
    if angka < 0:
        print("Faktorial hanya dapat dihitung untuk bilangan positif atau nol.")
    else:
        hasil = faktorial(angka)
        print(f"Faktorial dari {angka} adalah {hasil}")
except ValueError:
    print("Input tidak valid. Harap masukkan bilangan bulat.")
```

```python
# -*- coding: utf-8 -*-
"""nilai_tertinggi.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1cVQ3R2U1dUU7mkC7fIZY0h3D0M8yLdp4
"""

# Program untuk input nilai 5 siswa dan mencari nilai tertinggi

nilai_siswa = []

# Input nilai untuk 5 siswa
for i in range(5):
    while True:
        try:
            nilai = float(input(f"Masukkan nilai siswa ke-{i+1}: "))
            if 0 <= nilai <= 100:
                nilai_siswa.append(nilai)
                break
            else:
                print("Nilai harus antara 0 dan 100.")
        except ValueError:
            print("Input tidak valid. Masukkan angka.")

# Menentukan nilai tertinggi dan siswa keberapa
nilai_tertinggi = max(nilai_siswa)
siswa_tertinggi = nilai_siswa.index(nilai_tertinggi) + 1

print(f"\nNilai tertinggi adalah {nilai_tertinggi}, diperoleh oleh siswa ke-{siswa_tertinggi}.")
```

```python
# -*- coding: utf-8 -*-
"""hitung_diskon.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1IfyQ7vPf2PNUqYQ4vQHO9Xh2_J5BDFTX
"""

# Program untuk menghitung total bayar setelah diskon
def hitung_total_bayar(total_belanja):
    if total_belanja > 500000:
        diskon = total_belanja * 0.10
        print(f"Anda mendapatkan diskon sebesar 10%: Rp{diskon}")
    else:
        diskon = 0
        print("Anda tidak mendapatkan diskon.")
    total_bayar = total_belanja - diskon
    print(f"Total yang harus dibayar: Rp{total_bayar}")
    return total_bayar

# Input dari pengguna
try:
    total_belanja = float(input("Masukkan total belanja Anda (Rp): "))
    hitung_total_bayar(total_belanja)
except ValueError:
    print("Input tidak valid. Harap masukkan angka.")
```
